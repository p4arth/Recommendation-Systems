```python
import pandas as pd 
import numpy as np
```

```python
credits = pd.read_csv('/content/drive/MyDrive/tmdb_5000_credits.csv')
movies = pd.read_csv('/content/drive/MyDrive/tmdb_5000_movies.csv')
```

```python
merge = movies.merge(credits, on='title')
```

```python
merge.head()
```

```python
merge = merge[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
merge.head()
```

```python
merge.dropna(inplace=True)
```

```python
merge.iloc[0].genres
```

```python
import ast
def convert(obj):
  l = []
  for i in ast.literal_eval(obj):
    l.append(i['name'])
  return l 
```

```python
merge['genres']=merge['genres'].apply(convert)
```

```python
merge['keywords'] = merge['keywords'].apply(convert)
```

```python
merge.head()
```

```python
def convert2(obj):
  l = []
  counter = 0
  for i in ast.literal_eval(obj):
    if counter != 3:
      l.append(i['name'])
    else:
      break
  return l 
```

```python
merge['cast'] = merge['cast'].apply(convert2)
```

```python
def convert3(obj):
  l = []
  for i in ast.literal_eval(obj):
    if i['job'] == 'Director':
      l.append(i['name'])
      break
  return l 
```

```python
merge['crew'] = merge['crew'].apply(convert3)
```

```python
merge.head()
```

```python
merge['overview'] = merge['overview'].apply(lambda x: x.split())
```

```python
merge.head()
```

```python
merge['genres'] = merge['genres'].apply(lambda x: [i.replace(" ","") for i in x])
merge['keywords'] = merge['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
merge['cast'] =merge['cast'].apply(lambda x: [i.replace(" ","") for i in x])
merge['crew'] = merge['crew'].apply(lambda x: [i.replace(" ","") for i in x])
merge.head()
```

```python
merge['tags'] = merge['overview'] + merge['genres'] + merge['keywords'] + merge['cast'] + merge['crew']
```

```python
merge.head()
```

```python
df = merge[['movie_id','title','tags']]
df.head()
```

```python
df['tags']=df['tags'].apply(lambda x:" ".join(x))
```

```python
df.head()
```

```python
df['tags'][0]
```

```python
df['tags']=df['tags'].apply(lambda x: x.lower())
```

```python
df['tags'][1]
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y=[]
  for i in text.split():
    y.append(ps.stem(i))
  return " ".join(y)
```

```python
df['tags']=df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similar =  cosine_similarity(vectors)
```

```python
similar[0]
```

```python
def recommend(movie):
  movie_index = df[df['title'] == movie].index[0]
  distances = similar[movie_index]
  movies_list = sorted(list(enumerate(distances)), reverse=True,key=lambda x:x[1])[1:6]
  for i in movies_list:
    print(df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
